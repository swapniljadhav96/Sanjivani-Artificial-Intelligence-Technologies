```python
#find the data is normalise or not
#we can se the data is normali distrbuted
import numpy as np
import pylab
import scipy.stats as stats

measurements=np.random.normal(loc=20,scale=5 , size =100)   #generate  the random number
stats.probplot(measurements,dist='norm', plot=pylab)
pylab.show()
```
