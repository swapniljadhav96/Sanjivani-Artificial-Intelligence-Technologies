```python
#case study 2
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


#Load Q2.csv into a pandas dataFrame.
Q2 = pd.read_csv("C:/Data Science/3-Statistics/Q2_b.csv")
```

```python
# Scatterplot

Q2.plot(kind='scatter', x='SP', y="WT") ;
plt.show()

plt.hist(Q2.SP)

sns.FacetGrid(Q2, height=3) \
   .map(sns.distplot, "SP") \
   .add_legend();
plt.show();
```

```python
import numpy as np

counts,bin_edges = np.histogram(Q2['SP'], bins=10, density = True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)


counts, bin_edges = np.histogram(Q2['SP'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);

plt.show();

```

```python
# Inference
# PDF - Probablity Density function
# df will 0.04 to 0.10 from 5.0 to 25.0
# 

sns.boxplot(Q2.SP)
plt.show()
```

```python
sns.boxplot(Q2.WT)
plt.show()
```

```python
Q2.mean()
```

```python
print("\nStd-dev:");
print(np.std(Q2["SP"]))
print(np.std(Q2["WT"]))
```

```python
# skewness
from scipy.stats import skew
from scipy.stats import kurtosis



sp_lst=Q2['SP'].tolist()
print(skew(sp_lst))
print(kurtosis(sp_lst,axis=0,bias=True))


lst=Q2['SP'].tolist()
print(skew(lst))

print(kurtosis(lst,axis=0,bias=True))
```

```python
#Median, Quantiles, Percentiles, IQR.
print("\nMedians:")
print(np.median(Q2["SP"]))
#Median with an outlier
print(np.median(np.append(Q2["SP"],50)));
```
