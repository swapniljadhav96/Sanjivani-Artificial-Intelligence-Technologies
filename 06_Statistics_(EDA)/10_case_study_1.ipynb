```python
# Data Dictionary 

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np


df = pd.read_csv("C:/Data Science/3-Statistics/Q1_a.csv")
print(df)

```

```python
df.shape
# 81,5
```

```python
df.columns
#Index(['HP', 'MPG', 'VOL', 'SP', 'WT'], dtype='object')

```

```python
df.dtypes
#HP       int64
#MPG    float64
#VOL      int64
#SP     float64
#WT     float64
#dtype: object
#df.describe

```

```python
df.isnull()

```

```python
# Scatterplot

df.plot(kind='scatter', x='speed', y="dist") ;
plt.show()

plt.hist(df.speed)

sns.FacetGrid(df, height=4) \
   .map(sns.distplot, "speed") \
   .add_legend();
plt.show();

```

```python
import numpy as np

counts,bin_edges = np.histogram(df['speed'], bins=10, density = True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:], cdf)


counts, bin_edges = np.histogram(df['speed'], bins=20, 
                                 density = True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);

plt.show();

```

```python
# Inference
# PDF - Probablity Density function
# df will 0.04 to 0.10 from 5.0 to 25.0
# 

sns.boxplot(df.speed)
plt.show()

```

```python
sns.boxplot(df.dist)
plt.show()

```

```python
df.mean()
#Index    25.50
# speed    15.40
# dist     42.98
# dtype: float64

```

```python
print("\nStd-dev:");
print(np.std(df["speed"]))
print(np.std(df["dist"]))

```

```python
# skewness
from scipy.stats import skew
from scipy.stats import kurtosis



sp_lst=df['speed'].tolist()
print(skew(sp_lst))
print(kurtosis(sp_lst,axis=0,bias=True))


lst=df['dist'].tolist()
print(skew(lst))

print(kurtosis(lst,axis=0,bias=True))


```

```python
#Median, Quantiles, Percentiles, IQR.
print("\nMedians:")
print(np.median(df["speed"]))
#Median with an outlier
print(np.median(np.append(df["speed"],50)));
```
